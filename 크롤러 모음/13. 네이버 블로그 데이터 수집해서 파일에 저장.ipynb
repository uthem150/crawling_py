```python
print("=" *80)
print(" 파이썬 활용능력 1급 실기 시험 문제 ")
print(" 네이버 View 정보 수집 후 xls , csv 형식으로 저장하기")
print("=" *80)

#####################################################################################################################
# 과제명: 데이터 수집용 웹크롤러 만들기
#####################################################################################################################

#####################################################################################################################
# 반드시 아래 지시사항을 숙지한 후 시험에 응시하세요 !!
# 문제 : 아래의 Step 2의 38번 행과 Step 3의 45번행과 Step 5의 79번 행 view_list , Step 6의 115번 행의 코드를 완성하여 
# 웹 크롤러를 실행하여 데이터를 수집한 후 저장하여 결과물을 제출하세요.
# 시험에 대한 자세한 내용은 주어진 별도의 시험지를 참고하세요
# 제출 과제 누락시 실격처리가 되며 모든 책임은 응시자에게 있습니다.
#####################################################################################################################

#Step 1. 필요한 모듈과 라이브러리를 로딩하고 검색어를 입력 받습니다

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time
import sys       
import math
import pandas  as pd    

query_txt = "겨울여행"
cnt=50
page_cnt = math.ceil(cnt / 30)

fc_name='c:\\py_temp\\2311-00037_김도훈.csv'
fx_name='c:\\py_temp\\2311-00037_김도훈.xlsx'

#Step 2. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.
s = Service("c:/py_temp/chromedriver.exe")  # 문제 1. 크롬 드라이버의 경로를 올바로 지정하세요.
driver = webdriver.Chrome(service=s)

driver.get('http://www.naver.com')
time.sleep(2)

#Step 3. 네이버 검색창에 입력 받은 검색어를 넣고 검색한 후 "View" 선택
element = driver.find_element(By.ID,'query') # 문제 2. 이곳에 검색 창 정보를 지정하는 코드를 작성하세요
element.send_keys(query_txt)
element.send_keys("\n")

driver.find_element(By.LINK_TEXT,'VIEW').click( )
time.sleep(2)

# Step 4. 페이지 이동하기

no2 = [ ]           # 게시글 번호 컬럼
title2 = [ ]        # 게시물 제목 컬럼
content2 = [ ]     # 게시글 내용 컬럼
bdate2 = [ ]        # 작성 일자 컬럼
nick2 = [ ]         # 블로그 닉네임

no = 1

# 자동 스크롤다운 함수
def scroll_down(driver):
  driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
  time.sleep(5)

i = 1
while (i <= page_cnt):
      scroll_down(driver) 
      i += 1
    

print("\n")

# Step 5. 항목별 내용 추출하기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

view_list = soup.find('ul','lst_view _list_base').find_all('li') # 문제 3. 이곳에 게시글 목록을 전부 추출하는 코드를 작성하세요.

for i in view_list :
    
    try:
        contents = i.find('div','view_wrap').find_all('div')
    except :
        continue
    else :
    
        no2.append(no)                            # 게시물 번호 리스트에 추가
        print('1.번호:',no)

        #게시글 제목
        title= i.find('div','title_area').find('a').get_text()
        title2.append(title)
        print('2.게시글제목:' , title)

        #게시글 내용
        content = i.find('div','dsc_area').get_text()    
        content2.append(content)
        print('3.게시글내용:' , content)

        bdate = i.find('span','sub').get_text( )  # 작성일자
        bdate2.append(bdate)                     # 작성일자 리스트에 추가
        print('4.작성일자:',bdate)

        print("\n")

        no += 1

        if no > cnt :
            break

#Step 6. 출력 결과를 표(데이터 프레임) 형태로 만들어 파일로 저장하기

naver_blog = pd.DataFrame() # 문제 4.이 곳에 데이터프레임을 생성하는 코드를 작성하세요.
naver_blog['번호'] = no2
naver_blog['제목'] = title2
naver_blog['내용'] = content2
naver_blog['작성일자'] = bdate2

# csv 형태로 저장하기
naver_blog.to_csv(fc_name,encoding="utf-8-sig",index=False)
print(" csv 파일 저장 경로: %s" %fc_name) 

# 엑셀 형태로 저장하기
naver_blog.to_excel(fx_name , engine = 'openpyxl' , index=False)
print(" xls 파일 저장 경로: %s" %fx_name) 
```
